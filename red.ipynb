```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array ([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array ([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
#capa = tf.keras.layers.Dense(units=1, input_shape=[1])
#modelo = tf.keras.Sequential([capa])

oculta1 = tf.keras.layers.Dense(units=3, input_shape =[1])
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([oculta1,oculta2, salida])
```

```python
modelo.compile(
     #0.1 es cuanto peso de sesgo voy a darle, si es mucho peso ajustara mucho el margen de rror si le doy poco ajustara poco a poco.

    optimizer=tf.keras.optimizers.Adam(0.1),
    loss="mean_squared_error" #para la funcion de perdida usaremos este, esta funcion considera que una poca cantidad de errores grandes es peor que muchos pequeños
 )
```

```python
print("Comenzando el entrenamiento del robotboludo...")
historial = modelo.fit(celsius, fahrenheit, epochs=1200, verbose=False)
print("Modelo entrenado!")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de Pérdida")
plt.plot (historial.history ["loss"])
```

```python
print("hagamos una prediccion!")
resultado = modelo.predict([100.0])
print("el resultado es " + str(resultado) + " fahrenheit!")
```
